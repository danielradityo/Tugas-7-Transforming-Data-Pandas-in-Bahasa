```python
# 1. Import data FIFA menggunakan pandas
fifa = pd.read_csv('FIFA_dataset.csv')
fifa.head()
```

```python
# 2. Tampilkan hanya kolom 'Name', 'Age, 'Nationality'
fifa[['Name', 'Age', 'Nationality']]
```

```python
# 3. Tampilkan 5 baris teratas dan kolom 'Name', 'Overall', 'Potential', 'Release Clause' dengan menggunakan fungsi loc
fifa[['Name', 'Overall', 'Potential', 'Release Clause']].loc[:4]
```

```python
# 4. Tampilkan data hanya yang memiliki warganegara 'Germany!'
fifa[fifa['Nationality'] == 'Germany']
```

```python
# 5. Siapakah pemain yang memiliki warganegara 'France' dan berada pada Club 'Manchester United’
fifa[(fifa['Nationality'] == 'France') & (fifa['Club'] == 'Manchester United')]
```

```python
# 6. Tampilkan 'Name', 'Age', 'Overall', 'Wage', dan 'Release Clause' pada pemain dengan kewarganegaraan 'England' dan Club 'Manchester City'
fifa[['Name', 'Age', 'Overall', 'Wage', 'Release Clause']][(fifa['Nationality'] == 'England') & (fifa['Club'] == 'Manchester City')]
```

```python
# 7. Lakukan pengelompokkan Club ( groupby Club) dan tampilkan rata rata overall pemain
fifa.groupby(['Club']).agg({'Overall' : 'mean'})
```

```python
# 8. Lakukan pengelompokkan berdasarkan Nationality dan tampilkan jumlah pemain dan median umur
fifa.groupby(['Nationality']).agg({'Nationality' : 'count', 'Age' : 'median'})
```

```python
# 9. Membuat kolom baru bernama segment_player ', yang memuat status segmentasi pemain sbb:
    # a. 'High' jika overall pemain > 85
    # b. 'Mid' jika overall pemain antara 65 dan 85
    # c. 'Low' jika overall pemain dibawah 65
def segment_player(x):
    if x['Overall'] > 85:
        segment = 'High'
    elif 65 <= x['Overall'] <= 85:
        segment = 'Mid'
    else:
        segment = 'Low'
    return segment
        
fifa['segment_player'] = fifa.apply(lambda x: segment_player(x), axis = 1)
fifa
```

```python
# 10. Urutkan data berdasarkan Overall secara Descending
fifa.sort_values(by='Overall', ascending=False)
```
